```python
# 导入所需packages
import urllib3
from bs4 import BeautifulSoup as Soup
import pandas as pd
```

```python
# 初始化，确定爬取的页面范围
http = urllib3.PoolManager()
all_list =[]
stem = 'http://www.youjiaozhan.com/exhibitorlist/index.html?page='
pages = [stem + str(i) for i in range(1,9)]
```

```python
# 爬取每一家企业独立页面上的简介信息
def intro(temp_soup):
    
    paragraph = str()
    
    for i in temp_soup.find_all("div",class_='article-con'):
        for j in (i.find_all('p')):
            paragraph += j.get_text().strip()
            
    return paragraph
```

```python
# 综合独立简介信息与名称、联系方式
def get_info(dl):
    
    for i,v in dl.dt.a.attrs.items():
        temp = http.urlopen('get',v)
        temp_soup = Soup(temp.data.decode('utf-8'),'html.parser')
    
    temp_dict = {'ch_name':dl.h2.get_text(), 
                 'en_name':dl.h2.next_sibling.next_sibling.get_text(),
                 'telephone':dl.h2.next_sibling.next_sibling.next_sibling.next_sibling.get_text(),
                 'intro':intro(temp_soup)
                }
    
    print(temp_dict)  #如不需监测爬取过程，可将此行注释
    return temp_dict
```

```python
# 将一个公共页面上的所有企业信息按上述规则采集
def get_page(url):
    
    respond = http.urlopen('get',url)
    s = Soup(respond.data.decode('utf-8'),'html.parser')
    dl = s.dl
    temp_list = []
    
    while dl != None:
        temp_list.append(get_info(dl))
        dl=dl.next_sibling.next_sibling
        
    return temp_list
```

```python
# 爬取所有目标页面
for i in pages:
    all_list.append(get_page(i))
```

# 信息整理

```python
# 将list of list of dict转化为dataframe
flat_list = [item for sublist in all_list for item in sublist]
df = pd.DataFrame(flat_list)
```

```python
# 删除冗余信息
df.ch_name = df.ch_name.str.replace('中文：',"")
df.en_name = df.en_name.str.replace('英文：',"")
df.telephone = df.telephone.str.replace('电话：',"")
```

```python
# 重排表头，使用中文表头
df = df.reindex(['ch_name','en_name','telephone','intro'], axis=1)
df.columns = ['名称','英文名','电话','简介']
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
display(df)
```

```python
# 保存文件
df.to_excel('幼教展.xlsx')
```

Networkx Datacamp Test
=====

## Read in network file as G:

```python
file = 'ca-GrQc.txt'
```

```python
import networkx as nx
import nxviz as nv
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist(file)
```

```python
len(G.nodes())
```

```python
len(G.edges())
```

## Basic metrics

```python
%time
# Plot the degree distribution of the GitHub collaboration network
plt.hist(list(nx.degree_centrality(G).values()))
plt.show()
```

## Visualization of networks

```python
p = nv.ArcPlot(G)
p.draw()
plt.show()
```

```python
largest_ccs = sorted(nx.connected_component_subgraphs(G), key=lambda x: len(x))[-1]
h = nv.MatrixPlot(graph=largest_ccs)
h.draw()
plt.show()
```

```python
# generating benchmark networks
R = nx.erdos_renyi_graph(n=100,p=0.01)
```

```python
bench = nv.CircosPlot(R)
bench.draw()
plt.show()
```

```python
bench_m = nv.MatrixPlot(R)
bench_m.draw()
plt.show()
```

## Cliques

```python
# Calculate the maximal cliques in G: cliques
cliques = nx.find_cliques(G)
# Count and print the number of maximal cliques in G
print(len(list(cliques)))
```

```python
# Find the author(s) that are part of the largest maximal clique: largest_clique
largest_clique = sorted(nx.find_cliques(G), key=lambda x:len(x))[-1]

# Create the subgraph of the largest_clique: G_lc
G_lc = G.subgraph(largest_clique)

# Create the CircosPlot object: c
c = nv.CircosPlot(G_lc)

# Draw the CircosPlot to the screen
c.draw()
plt.show()
```
